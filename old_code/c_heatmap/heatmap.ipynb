```python
from copy import deepcopy
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv

filepath = 'muts/mutfeats.csv'
dataMatrix = pd.read_csv(filepath, index_col=0)

dataMatrix = dataMatrix[dataMatrix["class"] != 2]

df = pd.read_csv("muts/sample.csv")
namelist = df["Mutation"].tolist() + ['class']
dataMatrix = dataMatrix[namelist]

colorstore = {}
for index, row in dataMatrix.iterrows():
    if(row['class'] == 1.0):
        colorstore[index] = "green"
    else:
        colorstore[index] = "red"
colorstore
```

```python
#make it so we only see CIMP-High
dataMatrixPos = dataMatrix[dataMatrix['class'] == 1.0]

dataMatrix = dataMatrix.drop('class',1)
dataMatrixPos = dataMatrixPos.drop('class',1)
dataMatrix = dataMatrix.transpose()
dataMatrixPos = dataMatrixPos.transpose()

print("shape: ",dataMatrix.shape[0],dataMatrix.shape[1])
#dataMatrix = dataMatrix.loc[:, (dataMatrix != 0).any(axis=0)]

dataMatrix
```

```python
# just CIMP+
g = sns.clustermap(dataMatrixPos, figsize=(50,50), yticklabels=True, cmap='viridis')
for tick_label in g.ax_heatmap.axes.get_xticklabels():
    ttxt = tick_label.get_text()
    tick_label.set_color(colorstore[ttxt])
plt.savefig("muts/Cimppos.png")
plt.show()
```

```python
# all samples
g = sns.clustermap(dataMatrix, figsize=(30,20), yticklabels=True, cmap='viridis')
for tick_label in g.ax_heatmap.axes.get_xticklabels():
    ttxt = tick_label.get_text()
    tick_label.set_color(colorstore[ttxt])
plt.savefig("muts/all.png")
plt.show()
```
